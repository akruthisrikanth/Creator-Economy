```python
final_df = pd.concat([merged_df_spotify, merged_df_tiktok, merged_df_youtube])
final_df
```

```python
from datetime import date

def calculate_popularity_index(row, current_date):
    spotify_weight = 0.3
    youtube_weight = 0.4
    tiktok_weight = 0.3

    # Calculate the days difference between the current date and spotify_timestp
    timestamp = pd.to_datetime(row['Timestamp']).tz_localize(current_date.tz)
    days_difference = (current_date - timestamp).days

    decay_factor = 0.9
    time_decay = decay_factor ** days_difference

    # Calculate individual popularity scores for each factor
    if row['Spotify_Playlist_Total_Reach']!=0: 
        spotify_popularity_score = (row['Num_Spotify_Playlists'] / row['Spotify_Playlist_Total_Reach']) * row['Spotify_Streams']
    else:
        spotify_popularity_score = 0.0
    if row['YouTube_Playlist_Total_Reach']!= 0:
        youtube_popularity_score = (row['Num_YouTube_Playlists'] / row['YouTube_Playlist_Total_Reach']) * row['YouTube_Views']
    else:
        youtube_popularity_score = 0.0
    if row['TikTok_Counts']!= 0 :
        tiktok_popularity_score = (row['Num_TikTok_Videos'] / row['TikTok_Counts']) * row['TikTok_Counts']
    else:
        tiktok_popularity_score = 0.0

    # Combine the scores using the specified weights and apply time-based decay
    total_popularity_score = (time_decay * spotify_weight * spotify_popularity_score) + \
                            (time_decay * youtube_weight * youtube_popularity_score) + \
                            (time_decay * tiktok_weight * tiktok_popularity_score)

    return total_popularity_score

# current_date = pd.to_datetime('2023-07-20')
current_date = pd.to_datetime(date.today())

final_df['popularity_index'] = final_df.apply(calculate_popularity_index, args=(current_date,), axis=1)

artist_popularity_index = final_df.groupby('artist_id')['popularity_index'].mean().reset_index()

artist_popularity_index = artist_popularity_index.sort_values(by='popularity_index', ascending=False)

print(artist_popularity_index)
```
